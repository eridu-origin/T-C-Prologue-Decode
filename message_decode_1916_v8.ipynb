```python
# Decode a five bit binary word into an English Alphabet letter.
# There were 22 letters in the Elizabethan English Alphabet:
#   'I' = 'J'
#   'U' = 'V'


def decode_biliteral(biliteral):
  letter = {
  'aaaaa': 'a',
  'aaaab': 'b',
  'aaaba': 'c',
  'aaabb': 'd',
  'aabaa': 'e',
  'aabab': 'f',
  'aabba': 'g',
  'aabbb': 'h',
  'abaaa': 'i',
  'abaab': 'k',
  'ababa': 'l',
  'ababb': 'm',
  'abbaa': 'n',
  'abbab': 'o',
  'abbba': 'p',
  'abbbb': 'q',
  'baaaa': 'r',
  'baaab': 's',
  'baaba': 't',
  'baabb': 'v',
  'babaa': 'w',
  'babab': 'x',
  'babba': 'y',
  'babbb': 'z'
}
  return letter[biliteral]
  
```
